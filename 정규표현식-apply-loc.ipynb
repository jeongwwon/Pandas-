```python
import pandas as pd
movies=pd.read_csv('/content/drive/MyDrive/추천 시스템/data/ml-latest-small/movies.csv')
```

```python
movies['year'] = movies['title'].str.extract('(\(\d\d\d\d\))')
```

```python
movies['year']=movies['year'].str.extract('(\d\d\d\d)')
```

```python
movies['year'].unique()
```

```python
#apply는 해당 컬럼 for문을 돌며 함수 적용
movies['title'] = movies['title'].apply(lambda x:x.strip())
movies['apply'] = movies['title'].apply(lambda x:x[-5:-1])
movies['apply'].unique()
```

```python
errors=['r On', ' Roa', 'atso', 'imal', 'erso','lon ',
       'ligh', 'he O', 'osmo', ' Bab', 'ron ', 'irro']
```

```python
movies['apply'].isin(errors)
```

```python
#movies.loc[index,column] index==필터링 정보
movies.loc[movies['apply'].isin(errors),'apply']='2050'
```
